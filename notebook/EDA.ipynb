## Introduction
A Video Games Sale and Download Dataset contains data on video game sales, downloads, and user engagement across platforms. It includes details such as game title, genre, platform, release date, total sales (physical & digital), regional distribution, revenue, ratings, and in-game purchases. The dataset helps analyze market trends, player behavior, and sales performance, making it valuable for developers, publishers, and analysts in the gaming industry.

## Columns
* Rank – Position in sales ranking
* Name – Game title
* Platform – Console/PC system
* Year – Release year
* Genre – Type of game (Action, RPG, etc.)
* Publisher – Company that released the game
* NA_Sales – North America sales (in millions)
* EU_Sales – Europe sales (in millions)
* JP_Sales – Japan sales (in millions)
* Other_Sales – Sales in other regions (in millions)
* Global_Sales – Total worldwide sales (in millions)

```python
columns = list(df.columns)
```

```python
# Seperate Numeric and Categorical Columns
numeric_columns = [var for var in columns if df[var].dtype != 'O']
categorical_columns = [var for var in columns if df[var].dtype == 'O']
```

```python
style("Numeric Columns")
print(numeric_columns)
```

```python
# EDA on Year
sns.histplot(df['Year'], bins=20)
plt.title('Distribution of Year')
plt.show()
```

```python
plt.figure(figsize=(18, 9))

plot = 0
for var in numeric_columns:
    if var == 'Rank' or var == 'Year':
        continue
    else:
        plot += 1
        plt.subplot(2, 3, plot)
        sns.histplot(df[var], bins=20)
        plt.title(f'{var} Distribution')
        plt.xlabel(var)
        plt.ylabel('Count')
```

```python
# Skewness of Numeric Columns

style("Skewness of Numeric Columns")
for i in numeric_columns:
    if i == 'Rank' or i == 'Year':
        continue
    else:
        print(f"Skewness of {i} is: {round(df[i].skew(), 2)}")
```

```python
# Global Saales in easch Year
annually_global_sales = df.groupby('Year')['Global_Sales'].sum().sort_index(ascending=True)

sns.lineplot(x=annually_global_sales.index, y=annually_global_sales.values, color='red')
plt.title('Global Sales in each Year')
plt.xlabel('Year')
plt.ylabel('Global Sales(In Millions)')
plt.show()
```

```python
# Region wise sales

region_sales = df[['Other_Sales', 'EU_Sales', 'JP_Sales', 'NA_Sales']].sum(axis=0)

sns.barplot(x=region_sales.index, y=region_sales.values, palette='Set1')
plt.title('Region Wise Sales')
plt.xlabel('Regions')
plt.ylabel('Sales(In Millions)')
plt.show()
```

```python
# Year wise sales of regions

regions = ['Other_Sales', 'EU_Sales', 'JP_Sales', 'NA_Sales']
```

```python
plt.figure(figsize=(15, 10))

for i, region in enumerate(regions, 1):
    region_sales = df.groupby('Year')[region].sum().sort_index(ascending=True)

    plt.subplot(2, 2, i)
    sns.lineplot(x=region_sales.index, y=region_sales.values, color='green')
    plt.title(f'{region} in each Year')
    plt.xlabel('Year')
    plt.ylabel(region)
```

```python
# Top Game Publishers

publishers = df['Publisher'].value_counts().head(20)


plt.figure(figsize=(15,5))
sns.barplot(x=publishers.index, y=publishers.values,  palette='viridis')
plt.title('Top Publishers')
plt.xticks(rotation=90)
plt.xlabel('Publishers')
plt.ylabel('# of Released Games')
plt.show()
```

```python
# Top 20 Most Popular Games and there sales

popular_games = df.sort_values('Global_Sales', ascending=False).head(20)
```

```python
plt.figure(figsize=(15,8))

sns.barplot(orient='h', data=popular_games, x='Global_Sales', y='Name', hue='Publisher', palette='Set1')
plt.title('Top 20 Most Popular Games and there Global_Sales with Publishers')
plt.show()
```

```python
plt.figure(figsize=(15,8))

sns.barplot(orient='h', data=popular_games, x='Global_Sales', y='Name',hue='Year', palette='Set1')
plt.title('Top 20 Most Popular Games and there Global_Sales with Generation')
plt.show()
```

```python
# Get top 20 publishers
top_publishers = df['Publisher'].value_counts().head(20).index.to_list()
top_publishers
```

```python
def top_publisher_set(x):
    if x in top_publishers:
        return x
    else:
        return 'Others'
```

```python
df['Publisher'] = df['Publisher'].apply(top_publisher_set)
```

```python
plt.figure(figsize=(15,5))
sns.barplot(x='Publisher', y='Global_Sales', data=df, palette='Set2')
plt.title('Publishers vs Sales')
plt.xticks(rotation=90)
plt.xlabel('Publishers')
plt.ylabel('Global Sales')
plt.show()
```

```python
top_platfrorms = df['Platform'].value_counts().head(20).index.to_list()

def top_platfrorms_set(x):
    if x in top_platfrorms:
        return x
    else:
        return 'Others'
```

```python
df['Platform'] = df['Platform'].apply(top_platfrorms_set)
```

```python
plt.figure(figsize=(15,5))
sns.barplot(x='Platform', y='Global_Sales', data=df, palette='viridis')
plt.title('Platform vs Sales')
plt.xlabel('Platforms')
plt.ylabel('Sales')
plt.show()
```
